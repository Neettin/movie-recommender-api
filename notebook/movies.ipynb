```python
# Data Cleaning
df = df.dropna(subset=['title'])
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
# Now from selected columns ['genres'] lets extract only name not id using lambda function
df['genres'] = df['genres'].apply(lambda x : " ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
# for tagline
df['tagline'] = df['tagline'].fillna('')
```

```python
# for orginal_language
df['original_language'] = df['original_language'].fillna(df['original_language'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# combine 3 columns ['overview' + 'genres' + 'tagline'] for NLP vectorization
df['tags'] = df['overview'] + ' ' + df['genres'] + ' ' + df['tagline']
```

```python
# lets import stopwords, wordNetLemmatizer
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
# Download the stopwords and wordnet
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Let's create the object
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# Preprocessing the text
def preprocess_text(text):
  # Convert all in lower case
  text = str(text).lower()

  # Remove punctuation
  text = re.sub(r'[^\w\s]', '', text)

  # Tokenize or split the words
  words = text.split()

  # Remove stopwords
  words = [word for word in words if word not in stop_words]

  # lemmatization
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
# Applying above given function in tags column.
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
# Reset the index so that it should be in a sequence.
df = df.reset_index(drop = True)
```

```python
# Lets use indices for the title using pandas (series).
indices = pd.Series(df.index, index = df['title']).drop_duplicates()
indices
```

```python
# Importing Tf-Idf vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# create the object
tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1, 2), stop_words='english')
```

```python
# Vectorizing the tags
tfidf_matrix = tfidf.fit_transform(df['tags'])
tfidf_matrix
```

```python
# Import Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Lets create the fucntion for recommendation.
def recommend(title, n = 10):
    if title not in indices:
        return['Oops Sorry!. Movie Not Found.']
    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
```

```python
# testing
recommend('Avatar', 5)
```

```python
# Save the pickle file
import pickle

pickle.dump(tfidf_matrix, open('tfidf_matrix.pkl', 'wb'))

pickle.dump(indices, open('indices.pkl', 'wb'))

pickle.dump(tfidf, open('tfidf.pkl', 'wb'))

df.to_pickle('df.pkl')
```
